```python
import cv2
#cargamos los clasificadores
cars_classifier= cv2.CascadeClassifier('/content/drive/MyDrive/Computer Vision Masterclass/Cascades/cars.xml')
clock_calssifier= cv2.CascadeClassifier('/content/drive/MyDrive/Computer Vision Masterclass/Cascades/clocks.xml')
fullbody_classifier= cv2.CascadeClassifier('/content/drive/MyDrive/Computer Vision Masterclass/Cascades/fullbody.xml')
```

#Cars

```python
from google.colab.patches import cv2_imshow
image = cv2.imread('/content/drive/MyDrive/Computer Vision Masterclass/Images/car.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
car_detector= cars_classifier.detectMultiScale(gray_image, scaleFactor=1.04, minNeighbors=3)

for (x, y, w, h) in car_detector:
  cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)

cv2_imshow(image)
```

#Clocks

```python
clock_image = cv2.imread('/content/drive/MyDrive/Computer Vision Masterclass/Images/clock.jpg')

gray_clock_image = cv2.cvtColor(clock_image, cv2.COLOR_BGR2GRAY)

clock_detector= clock_calssifier.detectMultiScale(gray_clock_image, scaleFactor=1.03, minNeighbors=1)

for (x, y, w, h) in clock_detector:
  cv2.rectangle(clock_image, (x, y), (x + w, y + h), (0, 255, 0), 2)

cv2_imshow(clock_image)
```

#Fullbody

```python
fullbody_image = cv2.imread('/content/drive/MyDrive/Computer Vision Masterclass/Images/people3.jpg')

gray_fullbody_image = cv2.cvtColor(fullbody_image, cv2.COLOR_BGR2GRAY)

fullbody_detector= fullbody_classifier.detectMultiScale(gray_fullbody_image, scaleFactor=1.05, minNeighbors=5, minSize=(50, 50))



for (x, y, w, h) in fullbody_detector:
  cv2.rectangle(fullbody_image, (x, y), (x + w, y + h), (0, 255, 0), 2)

cv2_imshow(fullbody_image)
```
